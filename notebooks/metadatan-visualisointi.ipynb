```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')
```

```python
path = '/content/drive/Shareddrives/A-ICE/Data/AIS+ICE/data.csv'

df = pd.read_csv(path)
df.dropna(axis=0,inplace=True)

print(len(df))
df = df[df.DWT != 0]
print(len(df))
df = df[df.NRT != 0]
print(len(df))
df = df[df.GRT != 0]
print(len(df))
df = df[df.SPEED != 0]
print(len(df))
df = df[df.DRAFT != 0]
print(len(df))
df = df[df.POWER != 0]
print(len(df))
df = df[df.BREADTH != 0]
print(len(df))


```

```python
df_meta = df[['DWT', 'NRT', 'GRT', 'SPEED', 'DRAFT', 'POWER', 'BREADTH']]


for i in range(len(df_meta.columns)):
    for j in range(len(df_meta.columns)):
        plt.figure()
        plt.scatter(df_meta.iloc[:,i],df_meta.iloc[:,j])
        plt.xlabel(df_meta.columns[i])
        plt.ylabel(df_meta.columns[j])
        plt.show()
```
